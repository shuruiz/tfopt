```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Four-peaks

### Reading values from CSV files (output of the Jython code)

```python
# RHC
fp_rhc=pd.read_csv("fourpeaks_rhc.csv",header=None,sep=" ")
fp_rhc=fp_rhc.T
fp_rhc.columns=['Iterations','Fitness','Time']

# SA
fp_sa=pd.read_csv("fourpeaks_sa.csv",header=None,sep=" ")
fp_sa=fp_sa.T
fp_sa.columns=['Iterations','Fitness','Time']

# GA
fp_ga=pd.read_csv("fourpeaks_ga.csv",header=None,sep=" ")
fp_ga=fp_ga.T
fp_ga.columns=['Iterations','Fitness','Time']

```

### Plotting Compute Clock Time

```python
plt.figure(figsize=(8,5))
plt.semilogx(fp_rhc['Iterations'],fp_rhc['Time'],c='blue')
plt.semilogx(fp_sa['Iterations'],fp_sa['Time'],c='green')
plt.semilogx(fp_ga['Iterations'],fp_ga['Time'],c='red')

plt.grid(True)
plt.title("Compute clock time in $\mathbf{Four-peaks}$ problem",fontsize=16)
plt.xlabel("Number of iterations",fontsize=15)
plt.ylabel("Clock time (milliseconds)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(["Random Hill Climb","Simulated Annealing","Genetic Algorithm","MIMIC"],fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.loglog(fp_rhc['Iterations'],fp_rhc['Time'],c='blue')
plt.loglog(fp_sa['Iterations'],fp_sa['Time'],c='green')
plt.loglog(fp_ga['Iterations'],fp_ga['Time'],c='red')

plt.grid(True)
plt.title("Compute clock time in $\mathbf{Four-peaks}$ problem",fontsize=16)
plt.xlabel("Number of iterations",fontsize=15)
plt.ylabel("Clock time (milliseconds)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(["Random Hill Climb","Simulated Annealing","Genetic Algorithm","MIMIC"],fontsize=13)
plt.show()
```

### Plotting Fitness Score

```python
plt.figure(figsize=(8,5))
plt.semilogx(fp_rhc['Iterations'],fp_rhc['Fitness'],c='blue')
plt.semilogx(fp_sa['Iterations'],fp_sa['Fitness'],c='green')
plt.semilogx(fp_ga['Iterations'],fp_ga['Fitness'],c='red')

plt.grid(True)
plt.title("Fitness score in $\mathbf{Four-peaks}$ problem",fontsize=16)
plt.xlabel("Number of iterations",fontsize=15)
plt.ylabel("Fitness score",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(["Random Hill Climb","Simulated Annealing","Genetic Algorithm","MIMIC"],fontsize=13)
plt.show()
```

```python
plt.semilogy(fp_rhc['Fitness']/fp_rhc['Time'],c='blue')
plt.semilogy(fp_sa['Fitness']/fp_sa['Time'],c='green')
plt.semilogy(fp_ga['Fitness']/fp_ga['Time'],c='red')
# plt.semilogy(fp_mimic['Fitness']/fp_mimic['Time'],c='k')
plt.legend(["Random Hill Climb","Simulated Annealing","Genetic Algorithm","MIMIC"],loc=3)
plt.show()
```
